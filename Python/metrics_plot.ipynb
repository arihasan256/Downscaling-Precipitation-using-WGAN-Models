```python
import matplotlib.pyplot as plt
import numpy as np

# Data for UNET
unetpp_r2 = [0.9347733855247498, 0.9102043509483337, 0.9136451482772827]
unet_mae = [1.0468822, 1.121536, 1.1651478]
unet_ssim = [0.9851893, 0.9823439, 0.98354065]
unet_kge = [0.7507519747825829, 0.7554681585049983, 0.7493073053102766]

# Data for UNET++
unet_r2 = [0.9236043691635132, 0.9142459034919739, 0.9182026386260986]
unetpp_mae = [0.95887846, 1.1202095, 1.1699551]
unetpp_ssim = [0.9894516, 0.98239404, 0.9834918]
unetpp_kge = [0.7376872520530267, 0.7369946661358111, 0.7312422159718663]

# Dataset labels
datasets = ['Train', 'Validation', 'Test']

# Set width of bars
barWidth = 0.35
r1 = np.arange(len(datasets))
r2 = [x + barWidth for x in r1]

# Colors
unet_color = '#3498db'  # Blue
unetpp_color = '#e74c3c'  # Red

# Set up the figure and subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Perbandingan Metrics Model UNET dan UNET++', fontsize=16, fontweight='bold')

# Plot R2 Score (higher is better)
ax = axes[0, 0]
ax.bar(r1, unet_r2, width=barWidth, color=unet_color, edgecolor='white', label='UNET')
ax.bar(r2, unetpp_r2, width=barWidth, color=unetpp_color, edgecolor='white', label='UNET++')
ax.set_xlabel('Dataset', fontweight='bold')
ax.set_ylabel('Nilai', fontweight='bold')
ax.set_title('R² Score', fontweight='bold')
ax.set_xticks([r + barWidth/2 for r in range(len(datasets))])
ax.set_xticklabels(datasets)
ax.set_ylim(0.90, 0.94)  # Adjusted to zoom in on the differences
ax.grid(True, axis='y', linestyle='--', alpha=0.7)
for i, v in enumerate(unet_r2):
    ax.text(r1[i], v + 0.001, f'{v:.4f}', ha='center', va='bottom', fontsize=8)
for i, v in enumerate(unetpp_r2):
    ax.text(r2[i], v + 0.001, f'{v:.4f}', ha='center', va='bottom', fontsize=8)

# Highlight the better model with a marker for R2
for i in range(len(datasets)):
    if unet_r2[i] > unetpp_r2[i]:
        ax.scatter(r1[i], unet_r2[i] - 0.002, marker='^', color='green', s=80, zorder=3)
    else:
        ax.scatter(r2[i], unetpp_r2[i] - 0.002, marker='^', color='green', s=80, zorder=3)

# Plot Mean Absolute Error (lower is better)
ax = axes[0, 1]
ax.bar(r1, unet_mae, width=barWidth, color=unet_color, edgecolor='white', label='UNET')
ax.bar(r2, unetpp_mae, width=barWidth, color=unetpp_color, edgecolor='white', label='UNET++')
ax.set_xlabel('Dataset', fontweight='bold')
ax.set_ylabel('Nilai', fontweight='bold')
ax.set_title('Mean Absolute Error', fontweight='bold')
ax.set_xticks([r + barWidth/2 for r in range(len(datasets))])
ax.set_xticklabels(datasets)
ax.set_ylim(0.9, 1.2)  # Adjusted to zoom in on the differences
ax.grid(True, axis='y', linestyle='--', alpha=0.7)
for i, v in enumerate(unet_mae):
    ax.text(r1[i], v + 0.01, f'{v:.4f}', ha='center', va='bottom', fontsize=8)
for i, v in enumerate(unetpp_mae):
    ax.text(r2[i], v + 0.01, f'{v:.4f}', ha='center', va='bottom', fontsize=8)
    
# Highlight the better model with a marker for MAE
for i in range(len(datasets)):
    if unet_mae[i] < unetpp_mae[i]:
        ax.scatter(r1[i], unet_mae[i] - 0.02, marker='v', color='green', s=80, zorder=3)
    else:
        ax.scatter(r2[i], unetpp_mae[i] - 0.02, marker='v', color='green', s=80, zorder=3)

# Plot Structural Similarity Index (higher is better)
ax = axes[1, 0]
ax.bar(r1, unet_ssim, width=barWidth, color=unet_color, edgecolor='white', label='UNET')
ax.bar(r2, unetpp_ssim, width=barWidth, color=unetpp_color, edgecolor='white', label='UNET++')
ax.set_xlabel('Dataset', fontweight='bold')
ax.set_ylabel('Nilai', fontweight='bold')
ax.set_title('Structural Similarity Index', fontweight='bold')
ax.set_xticks([r + barWidth/2 for r in range(len(datasets))])
ax.set_xticklabels(datasets)
ax.set_ylim(0.982, 0.99)  # Adjusted to zoom in on the differences
ax.grid(True, axis='y', linestyle='--', alpha=0.7)
for i, v in enumerate(unet_ssim):
    ax.text(r1[i], v + 0.0002, f'{v:.4f}', ha='center', va='bottom', fontsize=8)
for i, v in enumerate(unetpp_ssim):
    ax.text(r2[i], v + 0.0002, f'{v:.4f}', ha='center', va='bottom', fontsize=8)

# Highlight the better model with a marker for SSIM
for i in range(len(datasets)):
    if unet_ssim[i] > unetpp_ssim[i]:
        ax.scatter(r1[i], unet_ssim[i] - 0.0004, marker='^', color='green', s=80, zorder=3)
    else:
        ax.scatter(r2[i], unetpp_ssim[i] - 0.0004, marker='^', color='green', s=80, zorder=3)

# Plot Kling-Gupta Efficiency (higher is better)
ax = axes[1, 1]
ax.bar(r1, unet_kge, width=barWidth, color=unet_color, edgecolor='white', label='UNET')
ax.bar(r2, unetpp_kge, width=barWidth, color=unetpp_color, edgecolor='white', label='UNET++')
ax.set_xlabel('Dataset', fontweight='bold')
ax.set_ylabel('Nilai', fontweight='bold')
ax.set_title('Kling-Gupta Efficiency', fontweight='bold')
ax.set_xticks([r + barWidth/2 for r in range(len(datasets))])
ax.set_xticklabels(datasets)
ax.set_ylim(0.73, 0.76)  # Adjusted to zoom in on the differences
ax.grid(True, axis='y', linestyle='--', alpha=0.7)
for i, v in enumerate(unet_kge):
    ax.text(r1[i], v + 0.001, f'{v:.4f}', ha='center', va='bottom', fontsize=8)
for i, v in enumerate(unetpp_kge):
    ax.text(r2[i], v + 0.001, f'{v:.4f}', ha='center', va='bottom', fontsize=8)

# Highlight the better model with a marker for KGE
for i in range(len(datasets)):
    if unet_kge[i] > unetpp_kge[i]:
        ax.scatter(r1[i], unet_kge[i] - 0.002, marker='^', color='green', s=80, zorder=3)
    else:
        ax.scatter(r2[i], unetpp_kge[i] - 0.002, marker='^', color='green', s=80, zorder=3)

# Add a legend at the bottom of the figure
handles, labels = axes[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', ncol=2, bbox_to_anchor=(0.5, 0.01), fontsize=12)

# Add annotation explaining the markers
plt.figtext(0.5, 0.05, '^ = better performance (higher is better) | v = better performance (lower is better)', 
            ha='center', fontsize=10, fontstyle='italic')

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(bottom=0.12)  # Add space at the bottom for the legend

# Save figure
plt.savefig('perbandingan_metrics_unet_unetpp.png', dpi=300, bbox_inches='tight')

# Show plot
plt.show()
```
